```r
# using subset to filter the data
demo2 <- subset(demo, marital == "Married")
head(demo2)
dim(demo2)
```

```r
# using subset with 2 filter conditions
demo2 <- subset(demo, marital == "Married" & age > 35)
head(demo2)
dim(demo2)
```

```r
# selecting a subset using subset with multiple filters and selecting particular columns
demo2 <- subset(demo, marital == "Married" & gender == "Male", select = c(1:5))
head(demo2)
dim(demo2)
```

```r
# selecting columns with subset and multiple filters
# select columns from 1 to 5 and 8th columns
demo2 <- subset(demo, marital == "Married" & gender == "Male", select = c(1:5, 8))
head(demo2)
```

```r
# Drop columns from 4 to 6 and 8 columns
demo2 <- subset(demo, marital == "Married" & gender == "Male", select = -c(4:6, 8))
head(demo2)
```
