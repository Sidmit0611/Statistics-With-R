```r
# select only females
demo2 <- demo[demo['gender'] == "Female", ]

# you can also write
#demo2 <- demo[demo$gender == "Female", ]
```

```r
head(demo2)
```

```r
# fetch the subject with income greater than 100
demo2 <- demo[demo['income'] > 100, ]
head(demo2)
dim(demo2)
```

```r
# if you want to keep only the variables 1, 3 and 7
demo2 <- demo[demo['income'] > 100, c(1, 3, 7)]
head(demo2)
```

```r
# what if you want to drop some variables lets say 6, 7 and 8
# then use ('-') sign with C
# here unlike python, upper index is included
# and index start from 1 instead of 0
demo2 <- demo[demo['income'] > 100, -c(6:8)]
head(demo2)
```

```r
head(demo)
```

```r
# select columns from 1 to 5
demo2 <- demo[c(1:5)]
head(demo2)
```

```r
# select the fields based on 2 filter
# in python we used &&(double and) instead of &(single and)
# now you can select the particular column using 'c'
demo2 <- demo[(demo['income'] > 100) & (demo['gender'] == "Female"), ]
head(demo2)
dim(demo2)
```
