```r
# read csv file 
demo <- read.csv("../input/demographics.csv")
```

```r
head(demo)
dim(demo)
```

```r
# we have categorical variable 'gender' which has male and female as its value
# changing male to '1'
# changing female to '2'
# we will make a new variable 'gender2'

demo$gender2[demo['gender'] == 'Male'] = '1'
```

```r
demo$gender2[demo['gender'] == "Female"] = '2'
```

```r
head(demo)
```

```r
# we can do the same type of decoding using the 'plyr' package using the 'revalue' function
require(plyr)
```

```r
demo$gender3 <- revalue(demo$gender, c("Male" = '1', "Female" = '2'))
head(demo)
```

## If a variable to recode is not a factor, we must convert it into a factor before recoding

```r
# to recode it into same variable (without creating a new one)
# we just use the same variable name in both sides of the revalue function
demo$gender <- revalue(demo$gender, c("Male" = '1', "Female" = '2'))
head(demo)
```
