```r
# read csv file 
demo <- read.csv("../input/demographics.csv")
```

```r
head(demo)
dim(demo)
```

```r
# we want to create a categorical variable such that if the income is lower than 200 then it is LowIncome
# and if income > 200 then it is HighIncome
demo$incat[demo['income'] < 200] = "Low Income"
demo$incat[demo['income'] >= 200] = 'High Income'
```

```r
head(demo)
```

```r
# now we want to create 3 groups by the variable income
# lower income < 150
# medium income 150-300
# and high income > 300
# so we will have two cut points: 150 and 300
# we will create a new variable incat2
# we have our cut points as -inf and +inf as anything lower than 150 and anything above 300 will have a cut
# as (-Inf, 150], (150, 300] and (300, Inf].
# open to the right '(' and close to the left ']'
# to get it conversely, use the option right = 'FALSE'

demo$incat2 = cut(demo$income, breaks = c(-Inf, 150, 300, Inf), labels = c('Low Income', 'Medium Income', 'High Income'), right = FALSE)
head(demo)
```
