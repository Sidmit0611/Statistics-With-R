```r
# read csv file 
demo <- read.csv("../input/demographics.csv")
```

```r
head(demo)
dim(demo)
```

```r
# sort the dataframe with income
# by default it is ascending, but you can change
demo <- demo[order(demo['income']), ]
```

```r
head(demo)
```

```r
# sorting the dataframe based on income but in increasing order
demo <- demo[order(-demo['income']), ]
```

```r
head(demo)
```

```r
# sorting the dataframe based on two variables
# income and age

demo <- demo[order(demo['income'], demo['age']), ]
head(demo)
```

```r
# sort by income ascending and age descending
demo <- demo[order(demo['income'], -demo['age']), ]
head(demo)
```
