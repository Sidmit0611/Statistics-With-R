# Getting the Statistical Indicators by groups or subsets

```r
# how to compute the main statistical indicators by groups and subsets
# import the package 
require(doBy)
```

```r
# define the function that generates the statistics
# N.B descStat is the command (in the doBy package) that computes the statistical indicators
func <- function(x) {descStat(x, na.rm = TRUE)} # na.rm is set to TRUE in order to ignore the missing values

# use the command SummaryBy
# so in this we are basically grouping by the data based on gender and the we are calculating the stats on income column
summaryBy(income~gender, data = demo, FUN = func)
```

```r
# get the statistical indicators for more than one variable
# indicators for the variables like income and age
summaryBy(income+age~gender, data = demo, FUN = func)
```

# Getting Statistical Indicators by Group with Stats

```r
# how to compute the main statistical indicators by groups or subsets
# with the package stats (command aggregate)
```

```r
# we will get some statistical indicators for the variable age
# by marital status
# compute the mean
aggregate(demo$age, by = list(demo$marital), FUN = mean)
```

```r
# this showed the mean age of person with married and unmarried criteria
```

```r
# compute the standard deviation
aggregate(demo$age, by = list(demo$marital), FUN = sd)
```

```r
# compute the median
aggregate(demo$age, by = list(demo$marital), FUN = median)
```

```r
# compute the variance
aggregate(demo$age, by = list(demo$marital), FUN = var)
```

```r
### This function is very useful when we combine the factor(categorical) categories 
# here you can see that we have more combinations by increasing the granularity
# suppose if someone asks to calculate the mean age of unmarried male
# you can easily get the results from this
aggregate(demo$age, by = list(demo$marital, demo$gender), FUN = mean)
```
